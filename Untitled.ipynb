```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt
import numpy as np
from scipy.interpolate import interp1d
```

```python
data = pd.read_csv('Medicion1Encoder.csv')

# Cada pulso es 1.08mm
PULSE_DISTANCE = 1.08 / 1000

times = data['Tiempos'].to_numpy()   
times = times / 1_000_000
distances = []
curr_distance = 0
for time in times:
    curr_distance += PULSE_DISTANCE
    distances.append(curr_distance)

distances = np.array(distances)
print(distances)
```

```python
velocities = []
for idx, time in enumerate(times):
    if idx != 0:
        curr_vel = (distances[idx] - distances[idx-1]) / (time - times[idx-1])  
    else:
        curr_vel = (distances[idx] - 0) / (time - 0)  
    velocities.append(curr_vel)
    
    
velocities = np.array(velocities)
print(velocities)
```

```python
print(times)
```

## PROCESAMOS VEL / DISTANCIA

```python
# Crear un grfico de seal discreta
plt.figure(figsize=(10, 5))
plt.stem(distances, velocities, basefmt=" ")
plt.title('Velocidad vs. Distancia (Seal Discreta)')
plt.xlabel('Distancia (m)')
plt.ylabel('Velocidad (m/s)')
plt.grid()

# Resaltar los ejes
ax = plt.gca()  # Obtener el eje actual
ax.spines['bottom'].set_linewidth(2)  # Grosor del eje x
ax.spines['left'].set_linewidth(2)    # Grosor del eje y

# Cambiar color de los ejes
ax.spines['bottom'].set_color('black')
ax.spines['left'].set_color('black')

plt.tight_layout()
plt.show()
```

```python
# Interpolacin
f_interp = interp1d(distances, velocities, kind='cubic')  # Interpolacin cbica
distance_interp = np.linspace(min(distances), max(distances), 100)  # Puntos interpolados
velocity_interp = f_interp(distance_interp)  # Velocidad interpolada
```

```python
# Crear grficos
plt.figure(figsize=(10, 5))

# Grfico de la seal original
plt.stem(distances, velocities, basefmt=" ", label='Datos Discretos', linefmt='r', markerfmt='ro')

# Grfico de la seal interpolada
plt.plot(distance_interp, velocity_interp, color='blue', label='Interpolacin Cbica')

# Configuracin del ttulo y ejes
plt.title('Interpolacin de Velocidad vs. Distancia')
plt.xlabel('Distancia (m)')
plt.ylabel('Velocidad (m/s)')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
